```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_excel("ESG-RV.xlsx","ESG_RV_Data")
df1 = pd.read_excel("ESG-RV.xlsx","CEMBI_ESG_RV")
df2 = pd.read_excel("ESG-RV.xlsx","EMBI_ESG_RV")
```

```python
x_col = "YTD ESG Score Change"
y_col = "%RC Average"
z_col = "Avg Rating"

t_col = "Trade"
s_col = "Combined ESG Score"

ratings = ["AA", "A", "BBB", "BB", "B", "CCC"]
colors = ['b', 'c', 'lawngreen', 'gold', 'darkorange', 'r']

fig = plt.figure(figsize=(16, 12))
ax = fig.add_subplot(111)
ax.axhline(y=0, xmin=-10.0, xmax=10.0, color='silver')
ax.axvline(x=0, ymin=-100.0, ymax=100.0, color='silver')

plt.scatter(df[x_col], df[y_col], alpha=0)
for i in range(len(ratings)):
    plt.scatter(df[x_col][df[z_col]==ratings[i]], df[y_col][df[z_col]==ratings[i]], marker='o', color=colors[i], label=ratings[i])

ax.text(df[x_col].max(), df[y_col].min(), 'UNATTRACTIVE', style='italic', 
        bbox={'facecolor': 'red', 'alpha': 0.5, 'pad': 10}, horizontalalignment='right')
ax.text(df[x_col].min(), df[y_col].max(), 'ATTRACTIVE', style='italic', 
        bbox={'facecolor': 'green', 'alpha': 0.5, 'pad': 10}, horizontalalignment='left')

plt.xlabel(x_col)
plt.ylabel(y_col)
plt.title('ESG Relative Value Chart')
plt.legend(loc='center left', bbox_to_anchor=(1, 0.5))

for i in range(df.shape[0]):
    ax.annotate('%s (%s)' % (df[t_col][i], df[s_col][i]), xy=(df[x_col][i],df[y_col][i]), textcoords='data')
```

```python
g_col = "BEMD SuperStrategy"
g_val = "CEMBI"

df_a = df[df[g_col]==g_val]
df_b = df1

s_col = "ESG Score"
t_col = "JPM Ticker"

x_col = "YTD ESG Score Change"
y_col = "%RC Average"
z_col = "Avg Rating"

ratings = ["AA", "A", "BBB", "BB", "B", "CCC", "CC", "C", "D", "NR"]
colors = ['b', 'c', 'lawngreen', 'gold', 'darkorange', 'r', 'firebrick', 'crimson', 'hotpink', 'darkorchid']

fig = plt.figure(figsize=(16, 12))
ax = fig.add_subplot(111)
ax.axhline(y=0, xmin=-10.0, xmax=10.0, color='silver')
ax.axvline(x=0, ymin=-100.0, ymax=100.0, color='silver')

plt.scatter(df_b[x_col], df_b[y_col], marker='^', color='silver', label=g_val, alpha=0.5)

for i in range(len(ratings)):
    plt.scatter(df_a[x_col][df_a[z_col]==ratings[i]], df_a[y_col][df_a[z_col]==ratings[i]], marker='o', color=colors[i], label=ratings[i])

ax.text(df_b[x_col].max(), df_b[y_col].min(), 'UNATTRACTIVE', style='italic', 
        bbox={'facecolor': 'red', 'alpha': 0.5, 'pad': 10}, horizontalalignment='right')
ax.text(df_b[x_col].min(), df_b[y_col].max(), 'ATTRACTIVE', style='italic', 
        bbox={'facecolor': 'green', 'alpha': 0.5, 'pad': 10}, horizontalalignment='left')

plt.xlabel(x_col)
plt.ylabel(y_col)
plt.title('ESG Relative Value Chart - '+g_val)
plt.legend(loc='center left', bbox_to_anchor=(1, 0.5))

for i in range(df_b.shape[0]):
    if abs(df_b[x_col][i]) > 0.4 and abs(df_b[y_col][i]) > 3:
        ax.annotate('%s (%s, %s)' % (df_b[t_col][i], df_b[z_col][i], df_b[s_col][i]), xy=(df_b[x_col][i],df_b[y_col][i]), textcoords='data')

for i in df_a.index:
    ax.annotate('%s (%s, %s)' % (df_a[t_col][i], df_a[z_col][i], df_a[s_col][i]), xy=(df_a[x_col][i],df_a[y_col][i]), textcoords='data')
```
